**6.1. Building a RAG system with LangChain**

**Retrieval & Generation**

```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_ollama import ChatOllama as LLM
# langchain library has been restructured recently. Use langchain_classic for older imports as mentioned below.
from langchain_classic.chains import RetrievalQA
```

```python
# configuration
vector_db_path = "/Users/mohammadne/Workspace/personal/ml_lab/courses/udemy_complete_genai/4.rag/1.rag_langchain/vector_db"
collection_name  = "document_collection"
```

```python
# loading the embedding model - default model
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# initialize ollama llm
llm = LLM(model="gemma3:1b", temperature=0.0)
```

```python
vector_store = Chroma(
    collection_name=collection_name,
    embedding_function=embedding,
    persist_directory=vector_db_path
)
```

```python
retriever = vector_store.as_retriever()
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "What does the document say about Adaptive Radiation?"
response = qa_chain.invoke({"query":query})
print(response["result"])
print("-"*80)
for source in response["source_documents"]:
    print(source.metadata)
```
