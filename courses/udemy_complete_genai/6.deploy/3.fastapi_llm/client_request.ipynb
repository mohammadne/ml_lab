```python
import requests

# we deploy on RunPod instead of AWS EC2 instances, the same installation scenarios
# like installing ollama, running model and host fastAPI server
url = "https://phmsibw0gl8zrd-8000.proxy.runpod.net/chat"

payload = {
    "model": "gemma2:2b",
    "messages": [
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "What is Machine Learning? Explain in short."}
    ]
}

response = requests.post(url, json=payload)

print(response.json())
```
