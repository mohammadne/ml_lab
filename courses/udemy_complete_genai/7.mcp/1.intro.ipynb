**MCP**

an open-source
standard that provides a unified, plug-and-play way for
AI models, particularly large language models (LLMs),
to connect with and use external tools, data sources,
and services. (by Anthropic)

Core Components:

- Host: The environment where the LLM runs (like an IDE or AI app). Itâ€™s the control center that connects users to tools and data through MCP.
- Client: The bridge between the host and servers. It handles all communication, turning requests and responses into a common MCP format.
- Server: The actual tool or data source exposed via MCP. It provides functions or information to the host in a standardized, easy-to-use way.
