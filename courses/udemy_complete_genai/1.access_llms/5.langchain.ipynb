**OpenAI**

```python
!pip install -q -U langchain-openai
```

```python
import os
from langchain_openai import ChatOpenAI as LLM

os.environ["OPENAI_API_KEY"] = "your_api_key"

llm = LLM(
    model="gpt-3.5-turbo",
    temperature=0,
)

response = llm.invoke("What is Machine Learning")

print(response.content)
```
